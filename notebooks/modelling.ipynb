```python
# read in data

metadata_files = glob.glob(f"{os.getcwd()}/../data/0_raw/app-store-reviews/google-play-store/**/META_*.json", recursive=True)
data_files = glob.glob(f"{os.getcwd()}/../data/0_raw/app-store-reviews/google-play-store/**/noLabel_*.json", recursive=True)

print(len(data_files))
print(len(metadata_files))
```

```python
data_dfs = []

for file in data_files[:1000]:
    f = open(file)
    data = json.load(f)
    data_dfs.append(pd.read_json(data, orient="records"))

raw_data_df = pd.concat(data_dfs)
raw_data_df.to_csv("../data/1_int/app-store-reviews/google_unioned_0_1000.csv", escapechar="\\")
```
